```python
#Exercise 0

#The following lines import the commands needed to plot functions
from numpy import arcsin, sin, cos, pi, linspace
import matplotlib.pyplot as plt

nI=1   #Index of refraction in incident medium
nT=1.5  #Index of refraction in transmitted medium

thetaI=linspace(0,pi,256, endpoint=True)  #This creates a list of 256 values for the incident theta  from 0 to pi

y=nI*sin(thetaI)*cos(thetaI)   #This calculates the values we will plot on the y-axis

plt.plot(thetaI,y, label='function y')  #Plot thetaI on the x-axis vs. y on the y-axis.  The label is the name that will appear in the legend.
plt.axis([0,pi,-0.6,0.6])  #Defines the limit of the axes.  The format of the command is [x_min, x_max, y_min, y_max]
plt.ylabel('Some generic function')  #Label for the y-axis
plt.xlabel('Incident angle - thetaI')  #Label for the x-axis
plt.grid(True)  #Include a grid
plt.legend(loc='upper left')  # Display a legend on the graph


plt.show()  #Display the graph
```

```python
#Exercise 1

from numpy import arcsin, sin, pi, linspace
import matplotlib.pyplot as plt


nI=1
nT=1.5

thetaI=linspace(0,pi/2,256, endpoint=True)

thetaT=arcsin(nI/nT*sin(thetaI))

plt.plot(thetaI,thetaT, label='thetaT')

plt.axis([0,pi/2,0,pi/2])  #Defines the limit of the axes.  The format of the command is [x_min, x_max, y_min, y_max]
plt.ylabel('Refracted angle - thetaT')  #Label for the y-axis
plt.xlabel('Incident angle - thetaI')  #Label for the x-axis
plt.grid(True)  #Include a grid
plt.legend(loc='upper left')  # Display a legend on the graph


plt.show()  #Display the graph
```

```python
#Exercise 2

from numpy import arcsin, sin,cos, pi, linspace
import matplotlib.pyplot as plt


nI=1
nT=1.5

thetaI=linspace(0,pi/2,256, endpoint=True)

thetaT=arcsin(nI/nT*sin(thetaI))



R_s = (nI*cos(thetaI)-nT*cos(thetaT))**2/(nI*cos(thetaI)+nT*cos(thetaT))**2

R_p = (nT*cos(thetaI)-nI*cos(thetaT))**2/(nT*cos(thetaI)+nI*cos(thetaT))**2

plt.plot(thetaI,R_s, label='$R_s$')
plt.plot(thetaI,R_p, label='$R_p$')


plt.axis([0,pi/2,0,1])  #Defines the limit of the axes.  The format of the command is [x_min, x_max, y_min, y_max]
plt.ylabel('Reflectance R')  #Label for the y-axis
plt.xlabel('Incident angle - thetaI')  #Label for the x-axis
plt.grid(True)  #Include a grid
plt.legend(loc='upper left')


plt.show()  #Display the graph
```

```python
#Exercise 5

from numpy import arcsin, sin,cos, pi, linspace
import matplotlib.pyplot as plt


nI=1
nT=1.5

thetaI=linspace(0,pi/2,256, endpoint=True)

thetaT=arcsin(nI/nT*sin(thetaI))



R_s1 = (nI*cos(thetaI)-nT*cos(thetaT))**2/(nI*cos(thetaI)+nT*cos(thetaT))**2

R_p1 = (nT*cos(thetaI)-nI*cos(thetaT))**2/(nT*cos(thetaI)+nI*cos(thetaT))**2

T_s = (nI*cos(thetaI)/(nT*cos(thetaT)))*(4*nT**2*cos(thetaT)**2)/(nT*cos(thetaT)+nI*cos(thetaI))**2  #At the second boundary, nI and nT switch roles as do thetaI and thetaT

T_p = (nI*cos(thetaI)/(nT*cos(thetaT)))*(4*nT**2*cos(thetaT)**2)/(nT*cos(thetaI)+nI*cos(thetaT))**2

R_s2=R_s1*T_s

R_p2=R_p1*T_p


plt.plot(thetaI,R_s2+R_s1, label='$R_{s\ Total}$')
plt.plot(thetaI,R_p2+R_p1, label='$R_{p\ Total}$')
plt.plot(thetaI,R_s1, label='$R_{s}$')
plt.plot(thetaI,R_p1, label='$R_{p}$')


plt.axis([0,pi/2,0,1])  #Defines the limit of the axes.  The format of the command is [x_min, x_max, y_min, y_max]
plt.ylabel('Reflectance R')  #Label for the y-axis
plt.xlabel('Incident angle - thetaI')  #Label for the x-axis
plt.grid(True)  #Include a grid
plt.legend(loc='upper left')

plt.show()  #Display the graph
```
